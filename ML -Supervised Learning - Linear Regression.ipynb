# Supervised Learning: Linear Regression


## Introduction to Supervised Learning

Supervised learning is a type of machine learning where the model is trained on a labeled dataset. This means that for every input in the training dataset, the corresponding output is provided. The goal of the supervised learning algorithm is to learn a mapping from inputs to outputs, which can then be used to predict the output for new inputs.

## Linear Regression Overview

Linear regression is one of the simplest and most widely used algorithms in supervised learning. It models the relationship between a dependent variable (target) and one or more independent variables (features) by fitting a linear equation to observed data.

## Linear Regression Equation

The equation for a simple linear regression model (with one feature) is:

$$ y = \beta_0 + \beta_1 x + \epsilon $$

where:
- \( y \) is the dependent variable.
- \( x \) is the independent variable.
- \( \beta_0 \) is the y-intercept.
- \( \beta_1 \) is the slope of the line.
- \( \epsilon \) is the error term.

For multiple linear regression (with multiple features), the equation is:

$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n + \epsilon $$

## Steps to Perform Linear Regression

1. **Collect Data**: Gather data containing the features and the target variable.
2. **Preprocess Data**: Handle missing values, encode categorical variables, and scale features if necessary.
3. **Split Data**: Divide the data into training and testing sets.
4. **Train the Model**: Fit the linear regression model to the training data.
5. **Evaluate the Model**: Assess the model's performance on the testing data.
6. **Make Predictions**: Use the model to make predictions on new data.

###  *Homoscedasticity*

The assumption of *homoscedasticity* (constant variance), is crucial to linear regression models. *Homoscedasticity* describes a situation in which the error term or variance or the "noise" or random disturbance in the relationship between the independent variables and the dependent variable is the same across all values of the independent variable. 

````python
plt.subplots(figsize = (12,8))
sns.residplot(x=data["enginesize"], y=data["price"])
````

From the above plot, we can tell the error variance across the true line is dispersed somewhat not uniformly, but in a funnel like shape. So, the assumption of the *homoscedasticity* is more likely not met.

### Normal distribution
The linear regression analysis requires the dependent variable, 'price', to be normally distributed. A histogram, box plot, or a Q-Q-Plot can check if the target variable is normally distributed. 

````python
import matplotlib.gridspec as gridspec
from scipy import stats

def plotting_3_chart(data, feature):
    # Set the style for the plots
    plt.style.use('fivethirtyeight')

    # Creating a customized chart and specifying figsize and layout
    fig = plt.figure(constrained_layout=True, figsize=(12, 8))
    # Creating a grid of 3 cols and 3 rows
    grid = gridspec.GridSpec(ncols=3, nrows=3, figure=fig)

    # Customizing the histogram grid
    ax1 = fig.add_subplot(grid[0, :2])
    # Set the title
    ax1.set_title('Histogram')
    # Plot the histogram
    sns.histplot(data.loc[:, feature], kde=True, ax=ax1)

    # Customizing the QQ_plot
    ax2 = fig.add_subplot(grid[1, :2])
    # Set the title
    ax2.set_title('QQ Plot')
    # Plotting the QQ Plot
    stats.probplot(data.loc[:, feature], plot=ax2)

    # Customizing the Box Plot
    ax3 = fig.add_subplot(grid[:, 2])
    # Set title
    ax3.set_title('Box Plot')
    # Plotting the box plot
    sns.boxplot(y=data.loc[:, feature], ax=ax3)
    
    plt.show()

# Assuming 'data' is the DataFrame and 'price' is the column to plot
plotting_3_chart(data, 'price')
````

These three charts above can tell us a lot about our target variable:

*   Our target variable, 'price' is not normally distributed
*   Our target variable is right-skewed
*   There are some outliers in the variable

# Understanding R-squared

R-squared, also known as the coefficient of determination, is a statistical measure used to evaluate the goodness of fit of a regression model. It indicates the proportion of the variance in the dependent variable that is predictable from the independent variables.

## Key Points About R-squared:

1. **Range**: R-squared values range from 0 to 1.
   - **0**: The model does not explain any of the variance in the dependent variable.
   - **1**: The model explains all the variance in the dependent variable.

2. **Interpretation**:
   - An R-squared value of 0.7, for example, means that 70% of the variance in the dependent variable is explained by the independent variables in the model.
   - Higher values indicate a better fit, meaning the model explains a greater portion of the variance.

3. **Formula**:
   \[
   R^2 = 1 - \frac{\text{SS}_{\text{res}}}{\text{SS}_{\text{tot}}}
   \]
   where:
   - \(\text{SS}_{\text{res}}\) is the sum of squares of the residuals (the difference between observed and predicted values).
   - \(\text{SS}_{\text{tot}}\) is the total sum of squares (the variance of the dependent variable).

4. **Limitations**:
   - **Doesn't Indicate Causation**: A high R-squared does not imply that the independent variables cause changes in the dependent variable.
   - **Overfitting**: In models with many predictors, R-squared can be artificially high, leading to overfitting. Adjusted R-squared is used to correct this by considering the number of predictors in the model.
   - **Comparing Models**: R-squared alone cannot be used to compare models with different dependent variables.

5. **Adjusted R-squared**:
   - Adjusted R-squared adjusts the R-squared value based on the number of predictors in the model. It is always lower than or equal to R-squared.
   - It is useful when comparing models with different numbers of predictors.

## Example:

Consider a simple linear regression where we predict house prices (dependent variable) based on square footage (independent variable).

1. **Calculate R-squared**:
   - Compute the predicted house prices using the regression model.
   - Calculate the residuals (observed price - predicted price) and sum of squares of residuals (\(\text{SS}_{\text{res}}\)).
   - Calculate the total sum of squares (\(\text{SS}_{\text{tot}}\)), which is the sum of squares of the difference between each observed price and the mean observed price.
   - Use the formula to compute R-squared.

2. **Interpretation**:
   - If R-squared is 0.85, it means 85% of the variance in house prices is explained by square footage.

R-squared is a valuable metric for evaluating the fit of a regression model, but it should be used in conjunction with other statistics and diagnostic tools to ensure a comprehensive understanding of model performance.


# Pipeline Object

We can also create a Pipeline object and apply a set of transforms sequentially. Then, we can apply linear regression. Data Pipelines simplify the steps of processing the data. We use the module Pipeline to create a pipeline. We also use StandardScaleras a step in our pipeline.

````python
steps=[('scaler', StandardScaler()), ('lm',  LinearRegression())]
pipe = Pipeline(steps=steps)
pipe.fit(X_train,y_train)

car_price_predictions = pipe.predict(X_test)
mse = mean_squared_error(y_test, car_price_predictions)
rmse = np.sqrt(mse)
rmse

r2_score(car_price_predictions, y_test)
````
